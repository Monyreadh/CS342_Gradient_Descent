```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
# Set the figure size for plots
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Read and preprocess the input data
data = pd.read_csv('data.csv')
X = data.iloc[:, 0].values
Y = data.iloc[:, 1].values

# Plot the input data
plt.scatter(X, Y)
plt.show()
```

```python
# Set the model parameters
learning_rate = 0.01
epochs = 1000

# Initialize the values of m and c
m = 0
c = 0

# Perform gradient descent
for i in range(epochs):
    # Compute the predicted values of Y
    Y_pred = m * X + c
    
    # Compute the gradients
    m_gradient = (-2/len(X)) * np.sum(X * (Y - Y_pred))
    c_gradient = (-2/len(X)) * np.sum(Y - Y_pred)
    
    # Update the values of m and c
    m -= learning_rate * m_gradient
    c -= learning_rate * c_gradient

# Print the final values of m and c
print(f'm: {m}, c: {c}')

# Make predictions using the final values of m and c
Y_pred = m * X + c

# Plot the input data and the regression line
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
# Calculate RMSE
rmse = sqrt(mean_squared_error(Y, Y_pred))
print(f'RMSE: {rmse}')

# Calculate MAE
mae = mean_absolute_error(Y, Y_pred)
print(f'MAE: {mae}')
```
